```python
import os

def list_folder_structure(start_path):
    for root, dirs, files in os.walk(start_path):
        level = root.replace(start_path, '').count(os.sep)
        indent = ' ' * 4 * level  # Indentation per folder level
        print(f'{indent}{os.path.basename(root)}/')
        sub_indent = ' ' * 4 * (level + 1)
        for file in files:
            print(f'{sub_indent}{file}')

# Replace '.' with the path to your file's directory
list_folder_structure('C:\dev\Smallfarming\farmadmin\templates\farmadmin\field_list.html')
```
